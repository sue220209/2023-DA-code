# 색상 및 각종 속성(스타일)지정

```python
import matplotlib.pyplot as plt

plt.plot([1,2,3,4], [10,20,30,40], color='skyblue',label='asc')
plt.plot([1,2,3,4], [40,30,20,10], color='pink',label='desc')
plt.title('legend')
plt.xlabel('x.axis')
plt.ylabel('y.axis')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.plot([1,2,3,4], [10,20,30,40], color='r',linestyle='--',label='dashed')
plt.plot([1,2,3,4], [40,30,20,10], color='b',ls=':',label='dotted')
plt.title('ls')
plt.xlabel('x.axis')
plt.ylabel('y.axis')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.plot([1,2,3,4], [10,20,30,40], color='r',linestyle='--',marker='d',label='dashed')
plt.plot([1,2,3,4], [40,30,20,10], color='b',ls=':',marker='^',label='dotted')
plt.title('marker')
plt.xlabel('x.axis')
plt.ylabel('y.axis')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.plot([1,2,3,4], [10,20,30,40], color='r',linestyle='--',marker='d',linewidth='4',label='dashed')
plt.plot([1,2,3,4], [40,30,20,10], color='b',ls=':',marker='^',label='dotted')
plt.title('linewidth')
plt.xlabel('x.axis')
plt.ylabel('y.axis')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 2 * np.pi, 200)
y = np.sin(x)
plt.plot(x, y)

plt.grid(True)
plt.rc('font', family='Malgun Gothic')
plt.rcParams['axes.unicode_minus']= False
plt.title('삼각함수')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

N = 8
y = np.zeros(N)

x1 = np.linspace(0, 10, N, endpoint=True)
x2 = np.linspace(0, 10, N, endpoint=False)
plt.plot(x1, y, 'o')
plt.plot(x2, y + 0.5, 'o')
plt.ylim([-0.5, 1])
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

N = 8
y = np.zeros(N)

x1 = np.linspace(0, 10, N, endpoint=True)
x2 = np.linspace(0, 10, N, endpoint=False)
plt.plot(x1, y, 'o-')
plt.plot(x2, y + 0.5, 'o:')
plt.ylim([-0.5, 1])
plt.show()
```
