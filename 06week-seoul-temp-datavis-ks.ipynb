```python
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv', encoding='cp949')
data = csv.reader(f)

header = next(data)

mean = []
low = []
high = []

for row in data:
    if (row[2] != '') and (row[3] != '') and (row[4] != ''):
        if(row[0].split('-')[1] =='12') and (row[0].split('-')[2] =='25'):
            mean.append(float(row[2]))
            low.append(float(row[3]))
            high.append(float(row[4]))

f.close()

plt.figure(figsize=(10,2))
plt.plot(mean, 'k')
plt.plot(low, 'b')
plt.plot(high, 'r')
plt.xlabel('Day(December 25)')
plt.ylabel('Temperature')
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv', encoding='cp949')
data = csv.reader(f)

header = next(data)

mean = []
low = []
high = []

for row in data:
    if (row[2] != '') and (row[3] != '') and (row[4] != ''):
        if(row[0].split('-')[1] =='12') and (row[0].split('-')[2] =='25'):
            mean.append(float(row[2]))
            low.append(float(row[3]))
            high.append(float(row[4]))

f.close()

plt.figure(figsize=(10,4))
plt.rc('font', family = 'Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False
plt.title('크리스마스의 기온 변화 그래프')
plt.plot(mean, 'k')
plt.plot(low, 'b')
plt.plot(high, 'r')
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt
f = open('seoul.csv', encoding='cp949')
data = csv.reader(f)

header = next(data)

temp_day = [] #1년 단위로 365일의 일별 평균 기온 데이터를 저장할 리스트 생성
temp_mean = [] #연도 별로, temp_day에 저장된 평균 기온 데이터의 평균 값을 저장할 리스트 생성
year = [] #연도 데이터를 저장할 리스트 생성 -> 그래프의 x축 데이터로 활용

for row in data:
    #1907년도에는 10월부터 기온 데이터가 기록되어 있어서, 평균 기온이 매우 낮음(분석에서 제외함)
    if int(row[0].split('-')[0]) > 1907:
        if row[2] != '':
            #1월 1일부터 12월 31일까지의 일별 평균 기온 데이터를 저장
            temp_day.append(float(row[2]))
            if (row[0].split('-')[1] == '12') and (row[0].split('-')[2] == '31'):
                year.append(int(row[0].split('-')[0])) #연도 데이터를 저장
                # 12월 31일이 되면 지금까지 temp_day에 저장했던 평균 기온 데이터의 평균 값을 계상
                temp_mean.append(sum(temp_day) / len(temp_day))
                temp_day = [] # 다음 년도의 평균 기온 데이터를 1월 1일부터 새롭게 저장해야하므로
                # temp_day 리스트를 [] 비어있는 리스트로 만들어 줌(이를 '초기화'라고 부름)

f.close()

plt.figure(figsize=(10, 4))
plt.rc('font', family ='Malgun Gothic')
plt.title('평균 기온 변화 그래프')
plt.plot(year, temp_mean, color = 'k', marker='d')
plt.xlabel('Year')
plt.ylabel('Average Temperature')
plt.grid()
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv', encoding='cp949')
data = csv.reader(f)
header = next(data)

#월별 데이터를 저장할 리스트  month(12개 생성)
month = [[], [], [], [], [], [], [], [], [], [], [], []]

for row in data:
    if row[4] != '':
        #월과 같은 번호의 인덱스에 월별 데이터 저장
        month[int(row[0].split('-')[1])-1].append(float(row[4]))

plt.boxplot(month)
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv', encoding='cp949')
data = csv.reader(f)
header = next(data)

day = []
for i in range(31):
    day.append( [] )

for row in data:
    if row[4] != '':
        if row[0].split('-')[1] == '01':
            #일과 같은 번호의 인덱스에 일별 데이터 저장
            day[int(row[0].split('-')[2])-1].append(float(row[4]))

plt.style.use('ggplot') #그래프 스타일 지정
plt.figure(figsize=(10, 5)) # 그래프 크기 수정
plt.boxplot(day, showfliers=False) # 이상치 표시 생략
plt.show()
```
