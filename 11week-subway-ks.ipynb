```python
import csv

f = open('subwayfee.csv', encoding='cp949')
data = csv.reader(f)

header = next(data)

max_people = [0] * 4
max_station = [''] * 4
label  = ['(유임승차)', '(유임하자)', '(무임승차)', '(무임하차)']

for row in data:
    for i in range(4, 8):
        row[i] = int(row[i].replace(',', ""))
        if row[i] > max_people[i-4]:
            max_people[i-4] = row[i]
            max_station[i-4] = row[3] + ' ' + row[1]

f.close()

for i in range(4):
    print(label[i] + ' ' + max_station[i] + ':', max_people[i], '명')
```

```python
import csv
import matplotlib.pyplot as plt

f = open('subwayfee.csv', encoding='cp949')
data = csv.reader(f)

header = next(data)

label  = ['(유임승차)', '(유임하자)', '(무임승차)', '(무임하차)']
color = ['skyblue', 'palegreen', 'pink', 'beige']

plt.rc('font', family='Malgun Gothic')
for row in data:
    for i in range(4, 8):
        row[i] = int(row[i].replace(',', ""))

    plt.title(row[3] + ' ' + row[1])
    plt.pie(row[4:8], labels=label, colors = color, autopct='%1.f%%')
    plt.show()
    break

f.close()
```

```python
import csv
import matplotlib.pyplot as plt

f = open('subwayfee.csv', encoding='cp949')
data = csv.reader(f)

header = next(data)

label  = ['(유임승차)', '(유임하자)', '(무임승차)', '(무임하차)']
color = ['skyblue', 'palegreen', 'pink', 'beige']

plt.rc('font', family='Malgun Gothic')

cnt = 0
for row in data:
    for i in range(4, 8):
        row[i] = int(row[i].replace(',', ""))

    plt.title(row[3] + ' ' + row[1])
    plt.pie(row[4:8], labels=label, colors = color, autopct='%1.f%%')
    plt.savefig(row[3] + ' ' + row[1] + '.png')
    plt.show()
    cnt += 1
    if cnt == 5:
        break

f.close()
```

## 내장 함수 map() : 기말고사 출제

```python
int('34')
```

```python
lst = ['34', '1999', '2786', '1000000']
result = map(int, lst)
list(result)
```

```python
pow(2, 3)
```

```python
# map에서 내장함수 활용
my_map = map(pow, [1,2,3], [2, 2, 2])
list(my_map)
```

```python
def myreplace(s):
    return s.replace(',', '')

lst = ['34,000', '1,999', '2,786', '7,000,000']
result = map(myreplace, lst)
list(result)
```

```python
int('10', 16)
```

```python
# map에서 내장 함수 int() 활용
list(map(int, [3, 5, 11]))
```

```python
# map에서 내장 함수 int() 활용
list(map(int, ['3', '5', '11']))
```

```python
list(map(int, ['11', '5', '11'], [2, 8, 16]))
```

```python
# map에서 중첩되게 map() 사용
s = [1, 5, 11]
list(map(str, s))
```

```python
list(map(int, map(str, [1, 5, 11]), [2, 8, 16]))
```

```python
# 람다(익명) 함수
a = lambda n: n+2
a(100)
```

```python
import csv
import matplotlib.pyplot as plt

f = open('subwaytime.csv', encoding='cp949')
data = csv.reader(f)

header = next(data)
next(data)

s_in = [0] * 24
s_out = [0] * 24

for row in data:
    for i in range(4, 52):
        row[i] = row[i].replace(',', "")

    row[4:] = map(int, row[4:])
    for i in range(24):
        s_in[i] = s_in[i] + row[2 * i + 4]
        s_out[i] = s_out[i] + row[2 * i + 5]

f.close()

plt.rc('font', family='Malgun Gothic')
plt.title('지하철 시간대별 승하차 인원 추이')
plt.plot(s_in, label='승차')
plt.plot(s_out, label='하차')
plt.legend()
plt.grid(True)
plt.xticks(range(24), range(4, 28))
plt.show()
```
